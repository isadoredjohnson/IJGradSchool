```python
#!/usr/bin/env python
# coding: utf-8
import pandas as pd
import numpy as np
df = pd.read_excel("/home/isadore/Downloads/D598 Data Set.xlsx")
df2 = df.groupby("Business State").mean(numeric_only=True)
df2
```

```python
print(df)
duplicates=df.duplicated()
print(df[duplicates])
#There are no duplicates in this dataset.
```

```python
States = df.groupby("Business State").agg({"mean", "median", "min", "max"})
print(States)
```

```python
FilteredSet = df[df['Debt to Equity']< 0]
print(FilteredSet)
```

```python
NewColumn = pd.DataFrame()
NewColumn['Debt-To-Income'] = df['Total Long-term Debt'].fillna(0) / df['Total Revenue'].fillna(1)
print(NewColumn)
```

```python
df = pd.concat([df, NewColumn[['Debt-To-Income']]], axis=1)
print(df)
```

```python
from matplotlib import pyplot as plt

top5states= df.sort_values(by='Total Revenue', ascending=False).head(5)

top5states.plot.bar(x='Business State', y='Total Revenue')
plt.title('Top 5 States with Highest Average Business Revenue')
plt.xlabel('State')
plt.ylabel('Average Revenue per Business')
plt.show()
```

```python
CTIndicators = df2.loc['Connecticut']
CTIndicators = CTIndicators.drop('Business ID')
print(CTIndicators)

CTIndicators.plot(kind='bar')
plt.title('Connecticut Business Indicators')
plt.xlabel('Indicators')
plt.ylabel('In 10 Millions')
plt.show
```

```python
plt.scatter(df['Total Long-term Debt'], df['Total Equity'])
plt.title('Long Term Debt on Equity')
plt.xlabel('Total Long Term Debt (in billions)')
plt.ylabel('Total Equity (in billions)')
plt.show
```
